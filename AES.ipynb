```python
# переводим текст любой длинны в матрицу состоящую из матриц 4х4, если плейн текст <= 16 то будет матрица в которой находится другая матрица 4х4
def text2_byte(text):
    txt = list(text.encode('utf-8').hex())
    temp = []
    for i in range(0, len(txt), 2):
        temp.append(txt[i]+txt[i+1])
    
    if len(temp)%16 == 0:
        size = len(temp)//16
    else:
        size = len(temp)//16 + 1
    
    for i in range(size*16 - len(temp)):
        temp.append(' '.encode('utf-8').hex())

    matrix = []
    for i in range(size):
        matrix.append([])
        for x in temp[i*16:i*16+16]:
            matrix[i].append(x)
    matrix = np.array(matrix)        
    for x in matrix:
        x = np.array(x).reshape(4,4)
    return matrix
```

```python
def shiftRows(matrix):
    shifted_matrix = []
    for x in range(len(matrix)):
        shifted_matrix.append(matrix[x].reshape(4,4))
    for x in range(len(shifted_matrix)): # ходим по матрице состоящей из матриц 4х4
        for i in range(4):
            shifted_matrix[x][i] = np.roll(shifted_matrix[x][i],-i)
    
    return shifted_matrix
```
